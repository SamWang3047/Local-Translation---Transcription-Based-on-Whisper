```python
from faster_whisper import WhisperModel

model_size = "turbo"

# Run on GPU with FP16
model = WhisperModel(model_size, device="cuda", compute_type="float16")

# or run on GPU with INT8
# model = WhisperModel(model_size, device="cuda", compute_type="int8_float16")
# or run on CPU with INT8
# model = WhisperModel(model_size, device="cpu", compute_type="int8")

mp3_file = "墨尔本大学 34-new.mp3"

segments, info = model.transcribe(mp3_file, beam_size=5)

print("Detected language '%s' with probability %f" % (info.language, info.language_probability))

# for segment in segments:
#     print("[%.2fs -> %.2fs] %s" % (segment.start, segment.end, segment.text))

# Get the txt filename
txt_file = mp3_file.rsplit('.', 1)[0] + ".txt"

# Open the txt file for writing
with open(txt_file, "w", encoding="utf-8") as f:
    for segment in segments:
        start_h = int(segment.start // 3600)
        start_m = int((segment.start % 3600) // 60)
        start_s = segment.start % 60
        end_h = int(segment.end // 3600)
        end_m = int((segment.end % 3600) // 60)
        end_s = segment.end % 60
        print("[%02d:%02d:%05.2f -> %02d:%02d:%05.2f] %s" % (start_h, start_m, start_s, end_h, end_m, end_s, segment.text))

        line = "[%02d:%02d:%05.2f -> %02d:%02d:%05.2f] %s\n" % (
            start_h, start_m, start_s, end_h, end_m, end_s, segment.text
        )
        f.write(line)

print(f"Saved to {txt_file}")
```

```python
# batch process
import os
from faster_whisper import WhisperModel

model_size = "turbo"

# Load the model once
model = WhisperModel(model_size, device="cuda", compute_type="float16")

# Your folder path
folder_path = "11142025"

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".mp3"):
        mp3_file = os.path.join(folder_path, filename)

        print(f"Processing {mp3_file}...")
        segments, info = model.transcribe(mp3_file, beam_size=5)

        print("Detected language '%s' with probability %f" % (info.language, info.language_probability))

        # Save output as a .txt file next to the .mp3
        txt_file = mp3_file.rsplit('.', 1)[0] + ".txt"
        
        with open(txt_file, "w", encoding="utf-8") as f:
            for segment in segments:
                start_h = int(segment.start // 3600)
                start_m = int((segment.start % 3600) // 60)
                start_s = segment.start % 60
                end_h = int(segment.end // 3600)
                end_m = int((segment.end % 3600) // 60)
                end_s = segment.end % 60
                line = "[%02d:%02d:%05.2f -> %02d:%02d:%05.2f] %s\n" % (
                    start_h, start_m, start_s, end_h, end_m, end_s, segment.text
                )
                f.write(line)

        print(f"Saved transcription to {txt_file}\n")

print("All files processed.")
```
